```r
pollutantmean <- function(directory,pollutant,id=1:332){
    pollu_v<-c()
    for(i in id){
        s_char<-nchar(i)
        if(s_char==1){
            i<-paste("00",i,sep="")
        }
        else if(s_char==2){
            i<-paste("0",i,sep="")
        }
        
        arch<- paste(directory,"/",i,".csv",sep="")
        csv<-read.csv(arch)
        pollu_v<-c(pollu_v,csv[[pollutant]])
    }
    mean(pollu_v,na.rm=TRUE)
    
}
```

```r
pollutantmean("specdata", "sulfate", 1:10)
```

```r
pollutantmean("specdata", "nitrate", 70:72)
```

```r
pollutantmean("specdata", "nitrate", 23)
```

```r
complete <- function(directory,id=1:332){
    comple<-data.frame()
    for(id in id){
        s_char<-nchar(id)
        ii<-id
        if(s_char==1){
            ii<-paste("00",id,sep="")
        }
        else if(s_char==2){
            ii<-paste("0",id,sep="")
        }
        
        arch<- paste(directory,"/",ii,".csv",sep="")
        csv<-read.csv(arch)
        nobs<-sum(!is.na(csv$sulfate) & !is.na(csv$nitrate))
        comple<-rbind(comple,data.frame(id,nobs))
    }
    print(comple)
    
}
```

```r
corr <- function(directory,threshold = 0){
    comple<- vector(mode = "numeric")
    for(id in 1:332){
        s_char<-nchar(id)
        ii<-id
        if(s_char==1){
            ii<-paste("00",id,sep="")
        }
        else if(s_char==2){
            ii<-paste("0",id,sep="")
        }
        
        arch<- paste(directory,"/",ii,".csv",sep="")
        csv<-read.csv(arch)
        nobs<-sum(!is.na(csv$sulfate) & !is.na(csv$nitrate))
        if(nobs>threshold){
            t<-0
            t<-cor(csv$sulfate,csv$nitrate, use = "complete.obs")
            comple<-c(comple,t)
        }
    }
    comple
    
}
```

```r
cr <- corr("specdata", 150)
head(cr)
summary(cr)
```
