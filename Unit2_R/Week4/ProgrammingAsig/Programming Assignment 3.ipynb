```r
rankhospital <- function(state, outcome, num = "best") {
## Read outcome data
    out <- read.csv("outcome-of-care-measures.csv", colClasses = "character")
    olt <- c("heart attack", "heart failure", "pneumonia")
    oltv <- c(11, 17, 23)
    olt.d<-data.frame(olt,oltv)
    out[, 11] <- as.numeric(out[, 11])
    out[, 17] <- as.numeric(out[, 17])
    out[, 23] <- as.numeric(out[, 23])    
## Check that state and outcome are valid
## Return hospital name in that state with the given rank
## 30-day death rate
    if(sum(out$State==state)>0){
        if(sum(outcomeslist==outcome)>0){
            pos<-olt.d[olt.d$olt==outcome,]$oltv
            out2<-out[out$State==state,]
            out2<-out2[order(out2[,pos],out2[,2]),]
            out2<-out2[!is.na(out2[,pos]),]
            if(num=="best"){
                num=1;
            }
            else if(num=="worst"){
                num=nrow(out2)
            }
            print(out2[order(out2[,pos],out2[,2]),][num,2])

        }
        else{
            cat(paste("Error in rankhospital(",state,", ",outcome,", ",num,") : invalid outcome",sep=""))
        }
        
    }
    else{
        cat(paste("Error in rankhospital(",state,", ",outcome,", ",num,") : invalid state",sep=""))
    }
}
rankhospital("TX", "pneumonia", 3)
```

```r
rankall <- function(outcome, num = "best") {
## Read outcome data
    out <- read.csv("outcome-of-care-measures.csv", colClasses = "character")
    olt <- c("heart attack", "heart failure", "pneumonia")
    oltv <- c(11, 17, 23)
    olt.d<-data.frame(olt,oltv)
    out[, 11] <- as.numeric(out[, 11])
    out[, 17] <- as.numeric(out[, 17])
    out[, 23] <- as.numeric(out[, 23])   
    x<-sort(unique(out$State))
    if(sum(outcomeslist==outcome)>0){
        hosps<-data.frame()
        for(i in x){
        ## Check that state and outcome are valid
            state <- i
            hospital <- ""

                    pos<-olt.d[olt.d$olt==outcome,]$oltv
                    out2<-out[out$State==state,]
                    out2<-out2[order(out2[,pos],out2[,2]),]
                    out2<-out2[!is.na(out2[,pos]),]
                    if(num=="best"){
                        num2=1;
                    }
                    else if(num=="worst"){
                        num2=nrow(out2)
                    }else if(num>0){
                        num2=num
                    }
                    hospital<-out2[order(out2[,pos],out2[,2]),][num2,2]

            hosps<-rbind(hosps,data.frame(hospital,state))

        }
        return(hosps)
    }
    else{
        cat(paste("Error in rankall(",outcome,", ",num,") : invalid outcome",sep=""))
    }
## For each state, find the hospital of the given rank
## Return a data frame with the hospital names and the
## (abbreviated) state name
}
```
